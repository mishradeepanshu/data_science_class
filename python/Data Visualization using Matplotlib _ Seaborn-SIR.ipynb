Data visualization is the graphic representation of data. It involves producing images that communicate relationships among the represented data to viewers. Visualizing data is an essential part of data analysis and machine learning. We'll use Python libraries Matplotlib and Seaborn to learn and apply some popular data visualization techniques in this session.

```python
import numpy as np
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# remove warnings
import warnings
warnings.filterwarnings('ignore')
```

## 1. Line Chart

The line chart is one of the simplest and most widely used data visualization techniques. A line chart displays information as a series of data points or markers connected by straight lines. You can customize the shape, size, color, and other aesthetic elements of the lines and markers for better visual clarity.

```python
tesla = [320, 335, 360, 350, 375, 380, 400]
```

```python
plt.plot(tesla)
```

### Customizing the X-axis

The X-axis of the plot currently shows list element indexes 0 to 5. The plot would be more informative if we could display the year for which we're plotting the data. We can do this by two arguments plt.plot.

```python
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
tesla = [320, 335, 360, 350, 375, 380, 400]
```

```python
plt.plot(years, tesla)
```

### Axis Labels
We can add labels to the axes to show what each axis represents using the plt.xlabel and plt.ylabel methods.

```python
plt.plot(years, tesla)
plt.title("Value of Tesla Stock Price for different years")
plt.xlabel('Years')
plt.ylabel('Value of the Stock per year')
```

```python
plt.figure(figsize = (16, 10))
plt.plot(years, tesla)
plt.xlabel('Years')
plt.ylabel('Value of the Tesla Stock per year')
```

### Plotting Multiple Lines
You can invoke the plt.plot function once for each line to plot multiple lines in the same graph.

```python
apple = [330, 332, 345, 350, 370, 385, 401]
years = [2015, 2016, 2017, 2018, 2019, 2020, 2021]
```

```python
plt.figure(figsize = (12, 7))
plt.plot(years, tesla)
plt.plot(years, apple)
plt.xlabel('Years')
plt.ylabel('Value of the Stock per year')
```

### Chart Title and Legend
To differentiate between multiple lines, we can include a legend within the graph using the plt.legend function. We can also set a title for the chart using the plt.title function.

```python
plt.figure(figsize = (12, 6))
plt.plot(years, tesla)
plt.plot(years, apple)

plt.xlabel('Years')
plt.ylabel('Value of the Stock per year')

plt.title('Comparison between 2 Stocks')
plt.legend(['Tesla', 'Apple'])
```

### Line Markers

We can also show markers for the data points on each line using the marker argument of plt.plot. Matplotlib provides many different markers, like a circle, cross, square, diamond, etc. 

You can find the full list of marker types here: https://matplotlib.org/3.1.1/api/markers_api.html .

Check out the documentation for plt.plot to learn more: https://matplotlib.org/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot .

```python
from matplotlib import markers
plt.figure(figsize = (12, 13))

plt.plot(years, tesla, marker = 'x')
plt.plot(years, apple, marker = 'o')

plt.xlabel('Years')
plt.ylabel('Value of the Stock per year')

plt.title('Comparison between 2 Stocks')
plt.legend(['Tesla', 'Apple'])
```

## 2. Scatter Plot

In a scatter plot, the values of 2 variables are plotted as points on a 2-dimensional grid. Additionally, you can also use a third variable to determine the size or color of the points. Let's try out an example.

```python
iris = pd.read_csv('iris.csv')
iris
```

```python
iris.columns
```

```python
iris['outputs'].unique()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(x = iris['sepal length in cm'], y = iris['sepal width in cm'])
```

### Adding Hues

Notice how the points in the above plot seem to form distinct clusters with some outliers. We can color the dots using the flower species as a hue. We can also make the points larger using the s argument.

```python
plt.figure(figsize=(13, 6))
sns.scatterplot(x = iris['sepal length in cm'], y = iris['sepal width in cm'], hue = iris['outputs'])
```

```python
plt.figure(figsize=(13, 6))
sns.scatterplot(x = iris['petal length in cm'], y = iris['petal width in cm'], hue = iris['outputs'])
```

### Customizing Seaborn Figures
Since Seaborn uses Matplotlib's plotting functions internally, we can use functions like plt.figure and plt.title to modify the figure.

```python
plt.figure(figsize=(13, 6))
plt.title('Petal Dimensions of Flower')

sns.scatterplot(x = iris['petal length in cm'], y = iris['petal width in cm'], hue = iris['outputs'], s = 80)
```

## 3. Pairplot
Seaborn also provides a helper function sns.pairplot to automatically plot several different charts for pairs of features within a dataframe.

```python
plt.figure(figsize=(13, 7))
plt.title('Dimensions of Flower')

sns.pairplot(iris, hue = 'outputs', size = 4)
```

## 4. Histogram
A histogram represents the distribution of a variable by creating bins (interval) along the range of values and showing vertical bars to indicate the number of observations in each bin.

```python
iris['sepal width in cm']
```

```python
plt.figure(figsize=(12, 7))
plt.title('Distribution of Sepal Width')

plt.hist(iris['sepal width in cm'])
```

### Controlling the size and number of bins
We can control the number of bins or the size of each one using the bins argument.

```python
plt.figure(figsize=(12, 7))
plt.title('Distribution of Sepal Width')

plt.hist(iris['sepal width in cm'], bins = 25)
```

```python
iris['outputs'].value_counts().values
```

## 5. Pie Chart
Pie charts show the size of items (called wedge) in one data series, proportional to the sum of the items. The data points in a pie chart are shown as a percentage of the whole pie. Matplotlib API has a pie() function that generates a pie diagram representing data in an array.

```python
data = pd.read_csv('titanic.csv')
data
```

```python
data['Pclass'].value_counts()
```

```python
plt.figure(figsize=(14, 7))

labels = ["3rd","1st","2nd"]
explode = [0, .1,.1]
colors = ['Red', 'Green', 'Yellow']
x = data['Pclass'].value_counts().values 

plt.title("Proportion of Passenger Class", size = 9)
plt.pie(x, explode, labels, colors, autopct = '%1.1f%%')
```

```python
data['Survived'].value_counts()
```

## 6. Count Plot
Countplot shows the counts of observations in each categorical bin using bars. A count plot can be thought of as a histogram across a categorical, instead of quantitative, variable. The basic API and options are identical to those for barplot() , so you can compare counts across nested variables.

```python
plt.figure(figsize=(14, 7))

sns.countplot(x = 'Pclass', data = data)
```

```python
plt.figure(figsize=(14, 7))

sns.countplot(x = 'Pclass', hue = 'Survived', data = data)
```

```python
sns.countplot(x = 'Sex', hue = 'Survived', data = data)
```

## 7. Bar Chart
Bar charts are quite similar to line charts, i.e., they show a sequence of values. However, a bar is shown for each value, rather than points connected by lines. We can use the plt.bar function to draw a bar chart.

```python
plt.figure(figsize=(14, 7))

sns.barplot(x = 'Pclass', y = 'Survived', hue = 'Sex', data = data)
```

*This graph demonstrates the proportions of survivability of passengers classes according to their genders*

*We can conclude from the first bar chart that the first class passenger class has relatively more survivors compared to the other classes, this may be due to the higher access of more advanced facilities provided to the the esteemed first class, like more lifeboats, life jackets, emergency doors.*

*We can also summarize that overall the male demographic was higher compared to the the females, ratio being 64% to the 36% females, and this is comparable to the survivability of the genders, where males passengers have survived more.*

**Passengers of the third class have the most number of population but still there are less survivor overall from their compared to the other classes this may be due to less number of survival facilities provided to them comapred to the elite ones.**

## 10. Heatmap

A heatmap is used to visualize 2-dimensional data like a matrix or a table using colors. The best way to understand it is by looking at an example. 

```python
housing = pd.read_csv("Housing (1).csv")
```

```python
housing.head()
```

```python
cor = housing.corr()
cor
```

```python
plt.figure(figsize = (12, 7))

cor = housing.corr()
sns.heatmap(cor, annot = True, cmap = 'Spectral_r')
plt.show()
```

Above plot shows the heatmap between diff rows and columns of the given Data chart.
Each label has a strong relationship with themself.
As the area and the number of bathrooms incerase, the price of house increases, indicating that they are somewhat proportional to each other.
